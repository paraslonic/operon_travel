```python
from ete3 import Tree
```

```python
t_core = Tree("studio/pdu/core_pdu_aligned.fasta.raxml.bestTree")
```

## compute for pdu

```python
t_pdu = Tree("studio/pdu/pdu.fasta.raxml.support")
out = t_pdu.robinson_foulds(t_core,unrooted_trees=True, min_support_t1 = 90, min_support_t2 = 80 )
print(out[0:2])
print(out[0]/out[1])
```

## compute for others

```python
import glob
```

```python
tree_files = glob.glob('tree_to_compare/pdu/trees/*.raxml.support')

for file in tree_files:
    tree = Tree(file)
    out = tree.robinson_foulds(t_core,unrooted_trees=True, min_support_t1 = 90, min_support_t2 = 80 )
    #print(out[0:2])
    print(out[0]/out[1])
```
